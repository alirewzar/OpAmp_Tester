```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data
data = [
    98.62237, 93.27894, 100.206, 109.8147, 93.32602, 95.45915, 90.45373,
    92.17799, 93.51049, 92.91423, 94.03044, 98.25391, 91.8692, 94.89078,
    92.55125, 92.67884, 92.49816, 89.24559, 95.29575, 94.3964, 96.40609,
    88.67457, 98.54455, 94.75988, 92.64531, 95.14894, 92.67446, 98.65138,
    94.51829, 93.38919
]

# Calculate mean and standard deviation
mean, std = np.mean(data), np.std(data)

# Plot histogram
plt.hist(data, bins=10, density=True, alpha=0.6, color='g')

# Plot normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)

title = "Fit results: mean = %.2f,  std = %.2f" % (mean, std)
plt.title(title)

plt.show()
```
